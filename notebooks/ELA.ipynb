```python
#Import Dataset
#===========================================================
import pandas as pd
URL = "../data/interim/healthcare-dataset-stroke-data.csv"
df = pd.read_csv(URL)
df.head()
```

```python
#transform data in binary
#================================================================================================================================
df['gender'] = df['gender'].replace({"Male":1,"Female":0,"Other":-1})
df = df[df['gender'] != -1]
df['ever_married'] = df['ever_married'].replace({"Yes":1,"No":0})
df['work_type'] = df['work_type'].replace({"Private":1,"Self-employed":2,"children":3,"Govt_job":4,"Never_worked":5})
df['Residence_type'] = df['Residence_type'].replace({"Urban":1,"Rural":0})
df['smoking_status'] = df['smoking_status'].replace({"never smoked":1,"Unknown":0,"formerly smoked":2,"smokes":3})
```

```python
df.dropna(inplace=True)
```

```python
#Saving of the transformed dataset
#=====================================================================
df.to_csv("../data/processed/healthcare-dataset-stroke-data-T.csv")
```

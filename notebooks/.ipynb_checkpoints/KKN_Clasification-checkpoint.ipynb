```python
#Import Dataset
#==============================================
import pandas as pd
URL = "../data/processed/healthcare-dataset-stroke-data-T.csv"
df = pd.read_csv(URL)
df.head()
```

```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values.ravel()
```

```python
plt.style.use('seaborn')
plt.figure(figsize = (10,10))
plt.scatter(X[:,0], X[:,1], c=y, marker= '*',s=100,edgecolors='black')
plt.show()
```

```python
#initiate Stratified K Fold
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=10)
skf.get_n_splits(X, y)  
Knn_scores  = {'accuracy':[],'precision':[],'recall':[]}
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score
#bucle
for train_index, test_index in skf.split(X, y):  
    #Model Split
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    #Train
    ModeloKNNC=model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    #Calculo de metricas para cada entrenamiento
    Knn_scores['precision'].append(precision_score(y_test,y_pred))
    Knn_scores['recall'].append(recall_score(y_test,y_pred))
    Knn_scores['accuracy'].append(accuracy_score(y_test,y_pred))
```

```python
#Mostramos resultado
from numpy import mean 
print('Overall precision_scores:',"{: .2%}".format(mean(Knn_scores['precision'])))
print('Overall recall_scores:',"{: .2%}".format(mean(Knn_scores['recall'])))
print('Overall accuracy_scores:',"{: .2%}".format(mean(Knn_scores['accuracy'])))
```
